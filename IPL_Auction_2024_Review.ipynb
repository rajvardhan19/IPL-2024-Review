```python
allRounder_2024 = csk_2024[csk_2024["Role"] == "All-rounder"].shape[0]
bowler_2024 = csk_2024[csk_2024["Role"] == "Bowler"].shape[0]
batsman_2024 = csk_2024[csk_2024["Role"] == "Batsman"].shape[0]
wicketkeeper_2024 = csk_2024[csk_2024["Role"] == "WK-Batsman"].shape[0]

print(f"Number of All-Rounders: {allRounder_2024}")
print(f"Number of Bowler: {bowler_2024}")
print(f"Number of Batsman: {batsman_2024}")
print(f"Number of Wicketkeeper: {wicketkeeper_2024}")
```

```python
compareData = pd.DataFrame({'2023':[allRounder_2023, bowler_2023, batsman_2023, wicketkeeper_2023], '2024':[allRounder_2024, bowler_2024, batsman_2024, wicketkeeper_2024]},index=['All-Rounders', 'Bowler', 'Batsman', 'Wicketkeeper'])

compareData
```

```python
for column in compareData.columns:
    plt.figure(figsize=(6, 6))
    plt.pie(compareData[column], labels=compareData.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Proportion in {column}')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
    plt.show()
```
