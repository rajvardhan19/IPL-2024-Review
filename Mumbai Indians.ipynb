```python
players = mi_soldPlayers["PLAYER"]
costs = mi_soldPlayers["PRICE PAID"]

plt.figure(figsize=(10, 6))
plt.plot(players, costs, marker='o', color='yellow', linestyle='-', linewidth=2, markersize=8)
plt.xlabel("PLAYER")
plt.ylabel('PRICE PAID')
plt.title("Player's Cost Variation")
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.gca().invert_yaxis()
plt.tight_layout()
plt.grid(True)  # Add grid lines
plt.show()
```

```python
country_counts = mi_2024['Nation'].value_counts()

# Plotting a bar graph
plt.figure(figsize=(8, 6))
country_counts.plot(kind='bar', color='yellow')
plt.xlabel('Country')
plt.ylabel('Frequency')
plt.title('Frequency of Countries')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
# Show the plot
plt.tight_layout()
plt.show()
```

```python
allRounder_2023 = mi_2023[mi_2023["TYPE"] == "ALL-ROUNDER"].shape[0]
bowler_2023 = mi_2023[mi_2023["TYPE"] == "BOWLER"].shape[0]
batsman_2023 = mi_2023[mi_2023["TYPE"] == "BATSMAN"].shape[0]
wicketkeeper_2023 = mi_2023[mi_2023["TYPE"] == "WICKETKEEPER"].shape[0]

print(f"Number of All-Rounders: {allRounder_2023}")
print(f"Number of Bowler: {bowler_2023}")
print(f"Number of Batsman: {batsman_2023}")
print(f"Number of Wicketkeeper: {wicketkeeper_2023}")
```

```python
allRounder_2024 = mi_2024[mi_2024["Role"] == "All-rounder"].shape[0]
bowler_2024 = mi_2024[mi_2024["Role"] == "Bowler"].shape[0]
batsman_2024 = mi_2024[mi_2024["Role"] == "Batsman"].shape[0]
wicketkeeper_2024 = mi_2024[mi_2024["Role"] == "WK-Batsman"].shape[0]

print(f"Number of All-Rounders: {allRounder_2024}")
print(f"Number of Bowler: {bowler_2024}")
print(f"Number of Batsman: {batsman_2024}")
print(f"Number of Wicketkeeper: {wicketkeeper_2024}")
```

```python
compareData = pd.DataFrame({'2023':[allRounder_2023, bowler_2023, batsman_2023, wicketkeeper_2023], '2024':[allRounder_2024, bowler_2024, batsman_2024, wicketkeeper_2024]},index=['All-Rounders', 'Bowler', 'Batsman', 'Wicketkeeper'])
compareData
```

```python
for column in compareData.columns:
    plt.figure(figsize=(6, 6))
    plt.pie(compareData[column], labels=compareData.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Proportion in {column}')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
    plt.show()
```
